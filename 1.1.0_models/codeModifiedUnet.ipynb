```python
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

import numpy as np
import cv2
from glob import glob
import tensorflow as tf
from tensorflow.keras.callbacks import (
    ModelCheckpoint, CSVLogger, ReduceLROnPlateau, EarlyStopping, TensorBoard
)
from tensorflow.keras.optimizers import Adam

# Fix seed
np.random.seed(42)
tf.random.set_seed(42)

H, W = 256, 256
smooth = 1e-15
```

```python
# ================= Metrics ================= #
def dice_coef(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    
    y_true = tf.reshape(y_true, [-1])  # Flatten tensor
    y_pred = tf.reshape(y_pred, [-1])
    
    intersection = tf.reduce_sum(y_true * y_pred)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) + smooth)

def iou(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    
    y_true = tf.reshape(y_true, [-1])
    y_pred = tf.reshape(y_pred, [-1])
    
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - intersection
    return (intersection + smooth) / (union + smooth)

def dice_loss(y_true, y_pred):
    return 1.0 - dice_coef(y_true, y_pred)

def combined_loss(y_true, y_pred):
    dice = dice_loss(y_true, y_pred)
    bce = tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)
    return 0.5 * dice + 0.5 * bce
```

```python
# ================= Data Loading ================= #
from sklearn.utils import shuffle
import albumentations as A

def load_data(path):
    x, y1, y2 = [], [], []
    patient_folders = sorted([p for p in glob(os.path.join(path, "*")) if os.path.isdir(p)])

    for patient in patient_folders:
        img_paths = sorted(glob(os.path.join(patient, "image", "*.png")))
        m1_paths = sorted(glob(os.path.join(patient, "mask", "*.png")))
        m2_paths = sorted(glob(os.path.join(patient, "mask1", "*.png")))
        n = min(len(img_paths), len(m1_paths), len(m2_paths))
        x.extend(img_paths[:n])
        y1.extend(m1_paths[:n])
        y2.extend(m2_paths[:n])

    return x, y1, y2

def read_image(path):
    path = path.decode()
    x = cv2.imread(path, cv2.IMREAD_COLOR)
    x = cv2.resize(x, (W, H))
    x = x / 255.0
    return x.astype(np.float32)

def read_mask(path1, path2):
    path1 = path1.decode()
    path2 = path2.decode()
    m1 = cv2.imread(path1, cv2.IMREAD_GRAYSCALE)
    m1 = cv2.resize(m1, (W, H))
    m1 = (m1 > 127).astype(np.float32)
    m2 = cv2.imread(path2, cv2.IMREAD_GRAYSCALE)
    m2 = cv2.resize(m2, (W, H))
    m2 = (m2 > 127).astype(np.float32)
    return m1[..., None], m2[..., None]

aug = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.ShiftScaleRotate(scale_limit=0.1, rotate_limit=15, p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
    A.GaussianBlur(blur_limit=(3, 7), p=0.3),
    A.ElasticTransform(alpha=120, sigma=120 * 0.05, alpha_affine=120 * 0.03, p=0.3)
])

def tf_parse(x, y1, y2, augment=False):
    def _parse(x, y1, y2):
        x = read_image(x)
        m1, m2 = read_mask(y1, y2)
        if augment:
            data = aug(image=x, masks=[m1, m2])
            x, m1, m2 = data["image"], data["masks"][0], data["masks"][1]
        return x, m1, m2

    x, m1, m2 = tf.numpy_function(_parse, [x, y1, y2], [tf.float32, tf.float32, tf.float32])
    x.set_shape([H, W, 3])
    m1.set_shape([H, W, 1])
    m2.set_shape([H, W, 1])
    return x, (m1, m2)

def tf_dataset(x, y1, y2, batch=4, augment=False):
    dataset = tf.data.Dataset.from_tensor_slices((x, y1, y2))
    dataset = dataset.map(lambda xx, yy1, yy2: tf_parse(xx, yy1, yy2, augment),
                          num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset
```

```python
# ================= Model Architecture ================= #
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Conv2DTranspose,
    concatenate, BatchNormalization, Activation
)
from tensorflow.keras.models import Model

def conv_block(inputs, num_filters):
    """Convolutional block: Conv2D -> BN -> ReLU -> Conv2D -> BN -> ReLU"""
    x = Conv2D(num_filters, 3, padding="same")(inputs)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Conv2D(num_filters, 3, padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    return x

def encoder_block(inputs, num_filters):
    """Encoder block: conv_block -> max_pooling"""
    s = conv_block(inputs, num_filters)
    p = MaxPooling2D((2, 2))(s)
    return s, p

def decoder_block(inputs, skip_features, num_filters):
    """Decoder block: upsampling -> concatenation -> conv_block"""
    x = Conv2DTranspose(num_filters, (2, 2), strides=2, padding="same")(inputs)
    x = concatenate([x, skip_features])
    x = conv_block(x, num_filters)
    return x

def build_unet(input_shape):
    """Build U-Net model with two output heads."""
    inputs = Input(input_shape)

    # Encoder path
    s1, p1 = encoder_block(inputs, 64)
    s2, p2 = encoder_block(p1, 128)
    s3, p3 = encoder_block(p2, 256)
    s4, p4 = encoder_block(p3, 512)

    # Bottleneck
    b1 = conv_block(p4, 1024)

    # Decoder path
    d1 = decoder_block(b1, s4, 512)
    d2 = decoder_block(d1, s3, 256)
    d3 = decoder_block(d2, s2, 128)
    d4 = decoder_block(d3, s1, 64)

    # Output heads
    # Each head is a 1x1 convolution with sigmoid activation
    output1 = Conv2D(1, 1, padding="same", activation="sigmoid", name="mask_elip")(d4)
    output2 = Conv2D(1, 1, padding="same", activation="sigmoid", name="mask_lesion")(d4)

    model = Model(inputs=inputs, outputs=[output1, output2], name="UNET")
    return model
```

```python
# ================= Data Splitting ================= #
from sklearn.model_selection import train_test_split
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
import os

dataset_path = "/home/lungscan/classification/data-mask/Lung-PET-CT-Dx-Mask"
X, Y1, Y2 = load_data(dataset_path)
train_x, temp_x, train_y1, temp_y1, train_y2, temp_y2 = train_test_split(
    X, Y1, Y2, test_size=0.3, random_state=42
)
valid_x, test_x, valid_y1, test_y1, valid_y2, test_y2 = train_test_split(
    temp_x, temp_y1, temp_y2, test_size=0.5, random_state=42
)

print("Train:", len(train_x))
print("Valid:", len(valid_x))
print("Test :", len(test_x))

train_dataset = tf_dataset(train_x, train_y1, train_y2, batch=8, augment=True)
valid_dataset = tf_dataset(valid_x, valid_y1, valid_y2, batch=8, augment=False)
test_dataset = tf_dataset(test_x, test_y1, test_y2, batch=8, augment=False)

# ================= Callbacks ================= #


# Tạo thư mục models nếu chưa tồn tại
os.makedirs("models", exist_ok=True)

# ================= Callbacks ================= #
callbacks = [
    ModelCheckpoint(
        filepath="models/best_model.keras",
        monitor="val_mask_elip_dice_coef",    # theo dõi dice coef của mask elip trên validation
        mode="max",                           # dice càng lớn càng tốt
        save_best_only=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=7,
        verbose=1
    ),
    CSVLogger("training_log.csv"),
    TensorBoard(log_dir="logs"),
    EarlyStopping(
        monitor="val_loss",
        patience=200,
        restore_best_weights=True,
        verbose=1
    )
]

# ================= Model Compilation and Training ================= #
model = build_unet((H, W, 3))
model.compile(
    optimizer=Adam(1e-4),
    loss={"mask_elip": combined_loss, "mask_lesion": combined_loss},
    metrics={
        "mask_elip": [dice_coef, iou, Precision(name="precision"), Recall(name="recall"), BinaryAccuracy(name="accuracy")],
        "mask_lesion": [dice_coef, iou, Precision(name="precision"), Recall(name="recall"), BinaryAccuracy(name="accuracy")]
    }
)

# Debug shapes
for x, (m1, m2) in train_dataset.take(1):
    print("Input shape:", x.shape)
    print("Mask 1 shape:", m1.shape)
    print("Mask 2 shape:", m2.shape)
    predictions = model(x, training=False)
    print("Prediction shapes:", [p.shape for p in predictions])

history = model.fit(
    train_dataset,
    validation_data=valid_dataset,
    epochs=200,
    callbacks=callbacks
)
```

```python
# ===== Evaluate on Test =====
results = model.evaluate(test_dataset)
print("Test metrics:", dict(zip(model.metrics_names, results)))
```

```python
# ================= Plot Learning Curves ================= #
import matplotlib.pyplot as plt

def plot_learning_curves(history):
    plt.figure(figsize=(15, 5))
    
    # Tổng loss (kết hợp cả mask_elip và mask_lesion)
    plt.subplot(1, 3, 1)
    plt.plot(history.history["loss"], label="Train Loss")
    plt.plot(history.history["val_loss"], label="Val Loss")
    plt.legend()
    plt.title("Total Loss")

    # Dice Coefficient cho mask_elip
    plt.subplot(1, 3, 2)
    plt.plot(history.history["mask_elip_dice_coef"], label="Train Dice (Elip)")
    plt.plot(history.history["val_mask_elip_dice_coef"], label="Val Dice (Elip)")
    plt.legend()
    plt.title("Dice Coefficient (Elip)")

    # IoU cho mask_elip
    plt.subplot(1, 3, 3)
    plt.plot(history.history["mask_elip_iou"], label="Train IoU (Elip)")
    plt.plot(history.history["val_mask_elip_iou"], label="Val IoU (Elip)")
    plt.legend()
    plt.title("IoU (Elip)")

    plt.show()

plot_learning_curves(history)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import label
import matplotlib.patheffects as pe

def show_one_image(img_path, model, img_size=(256, 256), threshold=0.5):
    # Đọc và chuẩn bị ảnh
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    if img is None:
        raise ValueError(f"Không thể đọc ảnh từ {img_path}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    H0, W0 = img.shape[:2]

    # Resize để predict
    img_resized = cv2.resize(img, img_size) / 255.0
    x = np.expand_dims(img_resized, axis=0).astype(np.float32)

    # Dự đoán xác suất (mô hình trả về list [mask_elip, mask_lesion])
    pred_elip, pred_lesion = model.predict(x, verbose=0)  # [batch, H, W, 1], [batch, H, W, 1]

    # Loại bỏ batch dimension và lấy channel đầu tiên
    pred_elip = pred_elip[0, :, :, 0]  # [H, W]
    pred_lesion = pred_lesion[0, :, :, 0]  # [H, W]

    # Resize về kích thước gốc
    elip_prob_map = cv2.resize(pred_elip, (W0, H0), interpolation=cv2.INTER_NEAREST)
    lesion_prob_map = cv2.resize(pred_lesion, (W0, H0), interpolation=cv2.INTER_NEAREST)

    # Tạo mask nhị phân
    elip_mask = (elip_prob_map > threshold).astype(np.uint8)
    lesion_mask = (lesion_prob_map > threshold).astype(np.uint8)

    # Hàm tính độ tin cậy và tách vùng
    def get_confidence_and_labels(mask, prob_map):
        labeled_array, num_features = label(mask)  # Tách các thành phần kết nối
        confidences = []
        for label_idx in range(1, num_features + 1):  # Bỏ nhãn 0 (nền)
            region_mask = (labeled_array == label_idx).astype(np.uint8)
            region_prob = prob_map[region_mask == 1]
            confidence = np.mean(region_prob) if region_prob.size > 0 else 0.0
            confidences.append((confidence, region_mask))
        return confidences

    # Tính độ tin cậy cho từng vùng
    elip_confidences = get_confidence_and_labels(elip_mask, elip_prob_map)
    lesion_confidences = get_confidence_and_labels(lesion_mask, lesion_prob_map)

    # Hiển thị
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    axes[0].imshow(img)
    axes[0].set_title("CT Gốc")

    # Hiển thị mask Elip (đen trắng)
    axes[1].imshow(elip_mask, cmap="gray")
    for confidence, region_mask in elip_confidences:
        if np.sum(region_mask) > 0:
            y, x = np.where(region_mask)
            top_y, center_x = np.min(y), int(np.mean(x))
            axes[1].text(center_x, top_y - 5, f"{confidence:.2f}",
                         color='yellow', ha='center', va='bottom', fontsize=8,
                         path_effects=[pe.withStroke(linewidth=1.5, foreground="black")])
    axes[1].set_title("Mask Elip")
    
    # Hiển thị mask Lesion (đen trắng)
    axes[2].imshow(lesion_mask, cmap="gray")
    for confidence, region_mask in lesion_confidences:
        if np.sum(region_mask) > 0:
            y, x = np.where(region_mask)
            top_y, center_x = np.min(y), int(np.mean(x))
            axes[2].text(center_x, top_y - 5, f"{confidence:.2f}",
                         color='yellow', ha='center', va='bottom', fontsize=8,
                         path_effects=[pe.withStroke(linewidth=1.5, foreground="black")])
    axes[2].set_title("Mask Lesion")

    for ax in axes:
        ax.axis("off")
    plt.tight_layout()
    plt.show()

# Ví dụ sử dụng
# img_path = "/kaggle/input/lung-mask/data/test/Lung-PET-CT-Dx-Mask/G0023/image/1-20.png"

# # Giả sử model đã được load với custom objects
# # from tensorflow.keras.models import load_model
# # model = load_model("model.h5", custom_objects={"dice_loss": dice_loss, "dice_coef": dice_coef, "iou": iou})
# show_one_image(img_path, model, img_size=(256, 256))
```
