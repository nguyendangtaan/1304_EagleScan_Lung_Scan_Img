```python
import os
import cv2
import numpy as np
import pydicom
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import albumentations as A
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.regularizers import l2

# ============================= CONFIG =============================
SEQUENCE_LENGTH = 10
TARGET_CENTER = 20
IMG_SIZE = 224
BATCH_SIZE = 8
EPOCHS_BINARY = 30 
EPOCHS_TRI = 200   
LUNG_PET_DIR = "/kaggle/input/lung-pet-ct-dx/Lung-PET-CT-Dx"
NORMAL_DIR = "/kaggle/input/normal/Normal"
target_names_binary = ["Normal", "Abnormal"]
target_names_tri = ["Adeno", "Small", "Squamous"]

# ========================= PREPROCESSING =========================
def apply_window(img_hu, ww, wl):
    img_min = wl - ww / 2
    img_max = wl + ww / 2
    img = np.clip(img_hu, img_min, img_max)
    return ((img - img_min) / ww * 255).astype(np.uint8)

def read_image_general(path):
    if path.lower().endswith(".dcm"):
        try:
            dicom = pydicom.dcmread(path)
            img = dicom.pixel_array.astype(np.float32)
            if hasattr(dicom, 'RescaleSlope'):
                img = img * dicom.RescaleSlope + dicom.RescaleIntercept
            return img
        except:
            return None
    else:
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        return img.astype(np.float32) if img is not None else None

# ========================= AUGMENTATION =========================
train_aug = A.ReplayCompose([
    A.HorizontalFlip(p=0.5),
    A.RandomRotate90(p=0.6),
    A.ShiftScaleRotate(shift_limit=0.2, scale_limit=0.2, rotate_limit=45, p=0.7),
    A.RandomBrightnessContrast(brightness_limit=0.4, contrast_limit=0.4, p=0.7),
    A.CLAHE(clip_limit=4.0, p=0.5), 
    A.Sharpen(p=0.3),             
    A.GaussNoise(var_limit=(10, 50), p=0.3), 
    A.CoarseDropout(max_holes=6, max_height=20, max_width=20, p=0.3),
], p=1.0)

# ========================= DATA GENERATOR =========================
class SequenceDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, x_files, y, batch_size, shuffle=True, aug=None):
        self.x_files = x_files
        self.y = y
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.aug = aug
        self.indices = np.arange(len(self.x_files))
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.x_files) / self.batch_size))

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __getitem__(self, index):
        idx = self.indices[index * self.batch_size:(index + 1) * self.batch_size]
        batch_x = [self.x_files[i] for i in idx]
        batch_y = self.y[idx]
        X = []
        for seq in batch_x:
            seq_imgs = []
            replay = None
            for i, (path, _) in enumerate(seq):
                img_hu = read_image_general(path)
                if img_hu is None:
                    img_hu = np.zeros((IMG_SIZE, IMG_SIZE))

                if path.lower().endswith(".dcm"):
                    img_l = apply_window(img_hu, 1500, -600)
                    medi = apply_window(img_hu, 350, 40)
                else:
                    img_gray = cv2.resize(img_hu.astype(np.uint8), (IMG_SIZE, IMG_SIZE))
                    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
                    img_l = clahe.apply(img_gray)
                    medi = clahe.apply(img_gray)

                img_l = cv2.resize(img_l, (IMG_SIZE, IMG_SIZE))
                medi = cv2.resize(medi, (IMG_SIZE, IMG_SIZE))

                if self.aug:
                    if i == 0:
                        res = self.aug(image=img_l)
                        replay = res["replay"]
                        img_aug = res["image"]
                    else:
                        img_aug = A.ReplayCompose.replay(replay, image=img_l)["image"]
                else:
                    img_aug = img_l

                final = cv2.merge([img_aug, medi, cv2.Canny(img_aug, 100, 200)])
                seq_imgs.append(resnet_preprocess(final.astype(np.float32)))

            X.append(np.stack(seq_imgs))
        return np.array(X), batch_y

# ========================= VISUALIZATION =========================
def plot_history(history, phase_name):
    metrics = ['loss', 'accuracy', 'auc']
    plt.figure(figsize=(18, 5))
    for i, m in enumerate(metrics):
        plt.subplot(1, 3, i+1)
        plt.plot(history.history[m], label='Train')
        plt.plot(history.history[f'val_{m}'], label='Val')
        plt.title(f'{phase_name} - {m.upper()}')
        plt.xlabel('Epochs')
        plt.legend()
    plt.tight_layout()
    plt.show()

# ========================= LOAD DATA =========================
def build_patient_data(directory, label_map):
    patients_info = []
    for folder in sorted(os.listdir(directory)):
        label_char = folder.split('-')[1][0] if '-' in folder else folder[0]
        if label_char not in label_map:
            continue
        path = os.path.join(directory, folder)
        all_series = []
        for root, _, files in os.walk(path):
            imgs = [os.path.join(root, f) for f in files if f.lower().endswith('.dcm')]
            if len(imgs) >= SEQUENCE_LENGTH:
                imgs.sort()
                all_series.append([(f, label_map[label_char]) for f in imgs])
        if all_series:
            patients_info.append((folder, label_map[label_char], all_series))
    return patients_info

def load_binary_data():
    X, y, p_ids = [], [], []
    label_map = {'A': 0, 'B': 1, 'G': 2}
    patients_info = build_patient_data(LUNG_PET_DIR, label_map)
    for p_id, label, series in patients_info:
        for slices in series:
            n = len(slices)
            target = 20 if n > 30 else (10 if n >= 10 else n)
            center = slices[(n - target) // 2 : (n - target) // 2 + target]
            for i in range(0, len(center) - SEQUENCE_LENGTH + 1, 2):
                X.append(center[i:i + SEQUENCE_LENGTH])
                y.append(1)  # Abnormal
                p_ids.append(p_id)

    all_subfolders = sorted([f for f in os.listdir(NORMAL_DIR) if os.path.isdir(os.path.join(NORMAL_DIR, f))])
    patient_to_subfolders = {}
    for sub in all_subfolders:
        patient_id = sub.split('_')[0]
        if patient_id not in patient_to_subfolders:
            patient_to_subfolders[patient_id] = []
        patient_to_subfolders[patient_id].append(sub)

    normal_count = 0
    for patient_id, sub_list in patient_to_subfolders.items():
        chosen_sub = sorted(sub_list)[0]
        sub_path = os.path.join(NORMAL_DIR, chosen_sub)
        imgs = []
        for root, _, files in os.walk(sub_path):
            for file in files:
                if file.lower().endswith(".png"):
                    imgs.append(os.path.join(root, file))
        imgs.sort()
        if len(imgs) < TARGET_CENTER:
            continue
        start = (len(imgs) - TARGET_CENTER) // 2
        center_imgs = imgs[start:start + TARGET_CENTER]
        for i in range(0, len(center_imgs) - SEQUENCE_LENGTH + 1, 2):
            seq = [(p, 3) for p in center_imgs[i:i + SEQUENCE_LENGTH]]
            X.append(seq)
            y.append(0)  # Normal
            p_ids.append(f"Normal_{patient_id}")
        normal_count += 1

    print(f"[INFO] Binary: {normal_count} bệnh nhân Normal + ung thư")
    y_cat = tf.keras.utils.to_categorical(y, num_classes=2)
    return np.array(X, dtype=object), y_cat, np.array(p_ids)

def load_tri_data():
    X, y, p_ids = [], [], []
    label_map = {'A': 0, 'B': 1, 'G': 2}
    patients_info = build_patient_data(LUNG_PET_DIR, label_map)
    for p_id, label, series in patients_info:
        for slices in series:
            n = len(slices)
            target = 20 if n > 30 else (10 if n >= 10 else n)
            center = slices[(n - target) // 2 : (n - target) // 2 + target]
            for i in range(0, len(center) - SEQUENCE_LENGTH + 1, 2):
                X.append(center[i:i + SEQUENCE_LENGTH])
                y.append(label)
                p_ids.append(p_id)
    y_cat = tf.keras.utils.to_categorical(y, num_classes=3)
    return np.array(X, dtype=object), y_cat, np.array(p_ids)

# ========================= GIAI ĐOẠN 1: BINARY =========================
print("=== GIAI ĐOẠN 1: BINARY CLASSIFICATION (Normal vs Abnormal) ===")
X_bin, y_bin, p_ids_bin = load_binary_data()

unique_pts_bin, first_idx_bin = np.unique(p_ids_bin, return_index=True)
patient_labels_bin = np.argmax(y_bin[first_idx_bin], axis=1)

p_train_bin, p_temp_bin, lbl_train_bin, lbl_temp_bin = train_test_split(unique_pts_bin, patient_labels_bin, test_size=0.3, stratify=patient_labels_bin, random_state=42)
p_val_bin, p_test_bin, lbl_val_bin, lbl_test_bin = train_test_split(p_temp_bin, lbl_temp_bin, test_size=0.333, stratify=lbl_temp_bin, random_state=42)

print("\n--- PHÂN BỐ BỆNH NHÂN BINARY ---")
print(f"Train: {len(p_train_bin)} bệnh nhân ({len(lbl_train_bin) - np.sum(lbl_train_bin)} Normal, {np.sum(lbl_train_bin)} Abnormal)")
print(f"Val: {len(p_val_bin)} bệnh nhân ({len(lbl_val_bin) - np.sum(lbl_val_bin)} Normal, {np.sum(lbl_val_bin)} Abnormal)")
print(f"Test: {len(p_test_bin)} bệnh nhân ({len(lbl_test_bin) - np.sum(lbl_test_bin)} Normal, {np.sum(lbl_test_bin)} Abnormal)")

idx_train_bin = np.isin(p_ids_bin, p_train_bin)
idx_val_bin = np.isin(p_ids_bin, p_val_bin)
idx_test_bin = np.isin(p_ids_bin, p_test_bin)

train_gen_bin = SequenceDataGenerator(X_bin[idx_train_bin], y_bin[idx_train_bin], BATCH_SIZE, aug=train_aug)
val_gen_bin = SequenceDataGenerator(X_bin[idx_val_bin], y_bin[idx_val_bin], BATCH_SIZE, shuffle=False)
test_gen_bin = SequenceDataGenerator(X_bin[idx_test_bin], y_bin[idx_test_bin], BATCH_SIZE, shuffle=False)

def attention_block(inputs):
    u = Dense(inputs.shape[-1], activation="tanh")(inputs)
    alpha = Softmax(axis=1)(Dense(1)(u))
    return Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1))([inputs, alpha])

def build_binary_model():
    inp = Input(shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3))
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base.trainable = True
    for layer in base.layers[:-20]:
        layer.trainable = False
    x = TimeDistributed(base)(inp)
    x = TimeDistributed(GlobalAveragePooling2D())(x)
    x = TimeDistributed(Dense(256, activation="relu"))(x)
    x = Bidirectional(GRU(32, return_sequences=True))(x)
    x = attention_block(LayerNormalization()(x))
    x = Dense(128, activation="relu", kernel_regularizer=l2(0.02))(x)
    out = Dense(2, activation="softmax")(Dropout(0.6)(x))
    model = Model(inp, out)
    model.compile(optimizer=AdamW(learning_rate=5e-5, weight_decay=1e-4),
                  loss=tf.keras.losses.CategoricalFocalCrossentropy(gamma=2.0),
                  metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
    return model

model_binary = build_binary_model()

callbacks_binary = [
    EarlyStopping(monitor='val_auc', mode='max', patience=20, restore_best_weights=True),
    ModelCheckpoint("binary_best.h5", save_best_only=True, monitor="val_auc", mode="max"),
    ReduceLROnPlateau(monitor="val_auc", mode='max', factor=0.5, patience=10)
]

history_binary = model_binary.fit(train_gen_bin, validation_data=val_gen_bin, epochs=EPOCHS_BINARY, callbacks=callbacks_binary)
plot_history(history_binary, "Binary Classification") # Vẽ biểu đồ

# Evaluation Binary
model_binary.load_weights("binary_best.h5")
test_preds_bin = model_binary.predict(test_gen_bin)

print("\n" + "="*60)
print("BINARY RESULT - SEQUENCE LEVEL")
y_true_seq_bin = np.argmax(y_bin[idx_test_bin], axis=1)
y_pred_seq_bin = np.argmax(test_preds_bin, axis=1)
print(classification_report(y_true_seq_bin, y_pred_seq_bin, target_names=target_names_binary, zero_division=0))

print("\nBINARY RESULT - PATIENT LEVEL (CONFIDENCE-WEIGHTED)")
y_true_pat_bin, y_pred_pat_bin = [], []
for p in p_test_bin:
    idx = np.where(p_ids_bin[idx_test_bin] == p)[0]
    if len(idx) > 0:
        p_preds = test_preds_bin[idx]
        # Thay đổi cơ chế voting sang Weighted: nhân với độ tự tin cực đại
        confidences = np.max(p_preds, axis=1, keepdims=True)
        weighted_avg = np.sum(p_preds * confidences, axis=0) / (np.sum(confidences) + 1e-7)
        y_pred_pat_bin.append(np.argmax(weighted_avg))
        y_true_pat_bin.append(patient_labels_bin[np.where(unique_pts_bin == p)[0][0]])

print(classification_report(y_true_pat_bin, y_pred_pat_bin, target_names=target_names_binary, zero_division=0))

# Confusion Matrix Binary
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(confusion_matrix(y_true_seq_bin, y_pred_seq_bin, labels=range(2)), annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_binary, yticklabels=target_names_binary, ax=ax[0])
ax[0].set_title("Binary - Sequence Level")
sns.heatmap(confusion_matrix(y_true_pat_bin, y_pred_pat_bin, labels=range(2)), annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_binary, yticklabels=target_names_binary, ax=ax[1])
ax[1].set_title("Binary - Patient Level")
plt.tight_layout()
plt.show()

# ========================= GIAI ĐOẠN 2: TRI-CLASS =========================
print("\n=== GIAI ĐOẠN 2: TRI-CLASS CLASSIFICATION (Adeno, Small, Squamous) ===")
X_tri, y_tri, p_ids_tri = load_tri_data()

unique_pts_tri, first_idx_tri = np.unique(p_ids_tri, return_index=True)
patient_labels_tri = np.argmax(y_tri[first_idx_tri], axis=1)

p_train_tri, p_temp_tri, lbl_train_tri, lbl_temp_tri = train_test_split(unique_pts_tri, patient_labels_tri, test_size=0.3, stratify=patient_labels_tri, random_state=42)
p_val_tri, p_test_tri, lbl_val_tri, lbl_test_tri = train_test_split(p_temp_tri, lbl_temp_tri, test_size=0.333, stratify=lbl_temp_tri, random_state=42)

print("\n--- PHÂN BỐ BỆNH NHÂN TRI-CLASS ---")
for name, lbl in zip(["Train", "Val", "Test"], [lbl_train_tri, lbl_val_tri, lbl_test_tri]):
    u, c = np.unique(lbl, return_counts=True)
    print(f"{name}: {len(lbl)} bệnh nhân")
    for i, count in zip(u, c):
        print(f"  - {target_names_tri[i]}: {count}")

idx_train_tri = np.isin(p_ids_tri, p_train_tri)
idx_val_tri = np.isin(p_ids_tri, p_val_tri)
idx_test_tri = np.isin(p_ids_tri, p_test_tri)

train_gen_tri = SequenceDataGenerator(X_tri[idx_train_tri], y_tri[idx_train_tri], BATCH_SIZE, aug=train_aug)
val_gen_tri = SequenceDataGenerator(X_tri[idx_val_tri], y_tri[idx_val_tri], BATCH_SIZE, shuffle=False)
test_gen_tri = SequenceDataGenerator(X_tri[idx_test_tri], y_tri[idx_test_tri], BATCH_SIZE, shuffle=False)

def build_tri_model():
    inp = Input(shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3))
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base.trainable = True
    for layer in base.layers[:-20]:
        layer.trainable = False
    x = TimeDistributed(base)(inp)
    x = TimeDistributed(GlobalAveragePooling2D())(x)
    x = TimeDistributed(Dense(256, activation="relu"))(x)
    x = Bidirectional(GRU(32, return_sequences=True))(x)
    x = attention_block(LayerNormalization()(x))
    x = Dense(128, activation="relu", kernel_regularizer=l2(0.02))(x)
    out = Dense(3, activation="softmax")(Dropout(0.6)(x))
    model = Model(inp, out)
    loss_fn = tf.keras.losses.CategoricalFocalCrossentropy(
        gamma=3.0,          # Tăng gamma để tập trung mẫu khó
        alpha=[0.3, 0.5, 0.2] # Ép trọng số lớp: Small(0.5) > Adeno(0.3) > Squamous(0.2)
    )
    model.compile(optimizer=AdamW(learning_rate=3e-5, weight_decay=1e-4),
                  loss=loss_fn,
                  metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
    return model

model_tri = build_tri_model()
model_tri.load_weights("binary_best.h5", by_name=True, skip_mismatch=True)

callbacks_tri = [
    EarlyStopping(monitor='val_auc', mode='max', patience=30, restore_best_weights=True),
    ModelCheckpoint("tri_best.h5", save_best_only=True, monitor="val_auc", mode="max"),
    ReduceLROnPlateau(monitor="val_auc", mode='max', factor=0.5, patience=15)
]

history_tri = model_tri.fit(train_gen_tri, validation_data=val_gen_tri, epochs=EPOCHS_TRI, callbacks=callbacks_tri)
plot_history(history_tri, "Tri-class Classification") # Vẽ biểu đồ

# Final evaluation tri-class
model_tri.load_weights("tri_best.h5")
test_preds_tri = model_tri.predict(test_gen_tri)

print("\n" + "="*60)
print("TRI-CLASS RESULT - SEQUENCE LEVEL")
y_true_seq_tri = np.argmax(y_tri[idx_test_tri], axis=1)
y_pred_seq_tri = np.argmax(test_preds_tri, axis=1)
print(classification_report(y_true_seq_tri, y_pred_seq_tri, target_names=target_names_tri, zero_division=0))

print("\nTRI-CLASS RESULT - PATIENT LEVEL (CONFIDENCE-WEIGHTED)")
y_true_pat_tri, y_pred_pat_tri = [], []
for p in p_test_tri:
    idx = np.where(p_ids_tri[idx_test_tri] == p)[0]
    if len(idx) > 0:
        p_preds = test_preds_tri[idx]
        # Thay đổi cơ chế voting sang Weighted: nhân với độ tự tin cực đại
        confidences = np.max(p_preds, axis=1, keepdims=True)
        weighted_avg = np.sum(p_preds * confidences, axis=0) / (np.sum(confidences) + 1e-7)
        y_pred_pat_tri.append(np.argmax(weighted_avg))
        y_true_pat_tri.append(patient_labels_tri[np.where(unique_pts_tri == p)[0][0]])

print(classification_report(y_true_pat_tri, y_pred_pat_tri, target_names=target_names_tri, zero_division=0))

# Confusion Matrix Tri-class
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.heatmap(confusion_matrix(y_true_seq_tri, y_pred_seq_tri, labels=range(3)), annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_tri, yticklabels=target_names_tri, ax=ax[0])
ax[0].set_title("Tri-class - Sequence Level")
sns.heatmap(confusion_matrix(y_true_pat_tri, y_pred_pat_tri, labels=range(3)), annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names_tri, yticklabels=target_names_tri, ax=ax[1])
ax[1].set_title("Tri-class - Patient Level")
plt.tight_layout()
plt.show()
```

```python
import os
import cv2
import numpy as np
import pydicom
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2

# ============================= CONFIG =============================
SEQUENCE_LENGTH = 10
IMG_SIZE = 224
BATCH_SIZE = 8
target_names_binary = ["Normal", "Abnormal"]
target_names_tri = ["Adeno", "Small", "Squamous"]

# ========================= PREPROCESSING THỐNG NHẤT CHO CẢ DICOM & PNG =========================
def apply_window(img_hu, ww, wl):
    img_min = wl - ww / 2
    img_max = wl + ww / 2
    img = np.clip(img_hu, img_min, img_max)
    return ((img - img_min) / (ww) * 255).astype(np.uint8)

def read_and_window_image(path):
    """
    Đọc ảnh và luôn áp dụng windowing chuẩn CT (Lung + Mediastinal)
    Dù là DICOM hay PNG → giảm bias format
    """
    if path.lower().endswith(".dcm"):
        try:
            dicom = pydicom.dcmread(path)
            img_hu = dicom.pixel_array.astype(np.float32)
            if hasattr(dicom, 'RescaleSlope') and hasattr(dicom, 'RescaleIntercept'):
                img_hu = img_hu * dicom.RescaleSlope + dicom.RescaleIntercept
        except:
            img_hu = np.zeros((512, 512))  # fallback
    else:
        # PNG: đọc grayscale, giả sử đã scale 0-255 → chuyển về HU gần đúng bằng cách giả lập range rộng
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            img_hu = np.zeros((512, 512))
        else:
            # Giả lập HU từ PNG: scale từ 0-255 → -1000 đến +1500 (phạm vi phổ biến của lung CT)
            img_hu = (img.astype(np.float32) / 255.0) * 2500 - 1000
    
    # Áp dụng windowing chuẩn cho mọi ảnh
    lung = apply_window(img_hu, ww=1500, wl=-600)   # Lung window
    medi = apply_window(img_hu, ww=350,  wl=40)     # Mediastinal window
    
    lung = cv2.resize(lung, (IMG_SIZE, IMG_SIZE))
    medi = cv2.resize(medi, (IMG_SIZE, IMG_SIZE))
    canny = cv2.Canny(lung, 100, 200)
    
    final = cv2.merge([lung, medi, canny])
    return resnet_preprocess(final.astype(np.float32))

# ========================= REBUILD MODEL (GIỐNG TRAIN) =========================
def attention_block(inputs):
    u = Dense(inputs.shape[-1], activation="tanh")(inputs)
    alpha = Softmax(axis=1)(Dense(1)(u))
    return Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1))([inputs, alpha])

def build_binary_model():
    inp = Input(shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3))
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base.trainable = True
    for layer in base.layers[:-20]:
        layer.trainable = False
    x = TimeDistributed(base)(inp)
    x = TimeDistributed(GlobalAveragePooling2D())(x)
    x = TimeDistributed(Dense(256, activation="relu"))(x)
    x = Bidirectional(GRU(32, return_sequences=True))(x)
    x = attention_block(LayerNormalization()(x))
    x = Dense(128, activation="relu", kernel_regularizer=l2(0.02))(x)
    out = Dense(2, activation="softmax")(Dropout(0.6)(x))
    return Model(inp, out)

def build_tri_model():
    inp = Input(shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3))
    base = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base.trainable = True
    for layer in base.layers[:-20]:
        layer.trainable = False
    x = TimeDistributed(base)(inp)
    x = TimeDistributed(GlobalAveragePooling2D())(x)
    x = TimeDistributed(Dense(256, activation="relu"))(x)
    x = Bidirectional(GRU(32, return_sequences=True))(x)
    x = attention_block(LayerNormalization()(x))
    x = Dense(128, activation="relu", kernel_regularizer=l2(0.02))(x)
    out = Dense(3, activation="softmax")(Dropout(0.6)(x))
    return Model(inp, out)

# ========================= INFERENCE GENERATOR =========================
class InferenceDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, sequences, batch_size):
        self.sequences = sequences
        self.batch_size = batch_size

    def __len__(self):
        return int(np.ceil(len(self.sequences) / self.batch_size))

    def __getitem__(self, index):
        batch_seq = self.sequences[index * self.batch_size:(index + 1) * self.batch_size]
        X = []
        for seq_paths in batch_seq:
            seq_imgs = [read_and_window_image(p) for p in seq_paths]
            X.append(np.stack(seq_imgs))
        return np.array(X)

# ========================= LOAD SEQUENCES =========================
def load_patient_sequences(patient_dir):
    imgs = []
    for root, _, files in os.walk(patient_dir):
        for f in files:
            if f.lower().endswith(('.dcm', '.png')):
                imgs.append(os.path.join(root, f))
    
    if len(imgs) == 0:
        print("[WARNING] Không tìm thấy ảnh")
        return None
    
    imgs.sort()
    n = len(imgs)
    if n > 30:
        target = 20
    elif n >= 10:
        target = 10
    else:
        print(f"[WARNING] Chỉ có {n} ảnh < 10 → bỏ qua")
        return None

    start_idx = (n - target) // 2
    center_imgs = imgs[start_idx:start_idx + target]

    sequences = []
    for i in range(0, len(center_imgs) - SEQUENCE_LENGTH + 1, 2):
        sequences.append(center_imgs[i:i + SEQUENCE_LENGTH])

    if not sequences:
        return None

    print(f"[INFO] Tải {len(sequences)} sequence ({n} ảnh gốc → {target} trung tâm)")
    return sequences

# ========================= PREDICT FUNCTION =========================
def predict_patient(patient_dir,
                    binary_weights_path="/kaggle/working/binary_best.h5",
                    tri_weights_path="/kaggle/working/tri_best.h5"):
    sequences = load_patient_sequences(patient_dir)
    if not sequences:
        return "Không đủ dữ liệu (cần ít nhất 10 ảnh)"

    gen = InferenceDataGenerator(sequences, BATCH_SIZE)

    print("Loading binary model weights...")
    model_binary = build_binary_model()
    model_binary.load_weights(binary_weights_path)

    preds_bin = model_binary.predict(gen, verbose=0)
    conf_bin = np.max(preds_bin, axis=1, keepdims=True)
    avg_bin = np.sum(preds_bin * conf_bin, axis=0) / (np.sum(conf_bin) + 1e-8)
    label_bin = np.argmax(avg_bin)
    score_bin = avg_bin[label_bin]

    result = f"🔍 BINARY: {target_names_binary[label_bin]} (Confidence: {score_bin:.4f})\n"

    if label_bin == 0:  # Normal
        return result + "→ Kết luận: Bình thường."

    print("Loading tri-class model weights...")
    model_tri = build_tri_model()
    model_tri.load_weights(tri_weights_path)

    preds_tri = model_tri.predict(gen, verbose=0)
    conf_tri = np.max(preds_tri, axis=1, keepdims=True)
    avg_tri = np.sum(preds_tri * conf_tri, axis=0) / (np.sum(conf_tri) + 1e-8)
    label_tri = np.argmax(avg_tri)
    score_tri = avg_tri[label_tri]

    result += f"🔬 TRI-CLASS: {target_names_tri[label_tri]} (Confidence: {score_tri:.4f})\n"
    result += f"   • Adenocarcinoma: {avg_tri[0]:.4f}\n"
    result += f"   • Small cell:     {avg_tri[1]:.4f}\n"
    result += f"   • Squamous cell:  {avg_tri[2]:.4f}"

    return result

# ========================= SỬ DỤNG =========================
print(predict_patient("/kaggle/input/lung-pet-ct-dx/Lung-PET-CT-Dx/Lung_Dx-A0044"))  
```
